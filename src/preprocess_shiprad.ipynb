# preprocess_shiprad

This notebook loads in the MOSAiC ARM ship-based radiation measurements and resamples them to 1 minute frequency.
It is recommended to run this notebook on the ARM Jupyterhub workbench: https://jupyterhub.arm.gov/

```python
import xarray as xr

import act
```

```python
# Download MOSAiC Shiprad data
# Set your username and token here. Get access token from here: https://adc.arm.gov/armlive/
username = '[your username here]'
token = '[your token here]'

# Set the datastream and start/enddates for the DL data
datastream = 'mosshipradS1.b1'
startdate = '2019-10-15T00:00:00'
enddate = '2020-09-18T23:59:59'

# Use ACT to easily download the data.  Watch for the data citation!  Show some support
# for ARM's instrument experts and cite their data if you use it in a publication
in_files = act.discovery.download_arm_data(username, token, datastream, startdate, enddate)
ds_in = act.io.read_arm_netcdf(in_files)
```

```python
ds_in
```

```python
da_lw = ds_in['downwelling_longwave'].where(ds_in['qc_downwelling_longwave'] == 0)

ds_subset = xr.merge([ds_in[['spn1_total_corr', 'lat', 'lon']], da_lw])
ds_subset.load()
```

```python
ds_out = ds_subset.resample({'time':'1min'}).mean()
```

```python
ds_out.to_netcdf('mosshipradS1.b1.subset.20191015_20200918.nc')
```
